## MEDIA ANNUA PER MKT ETF

```python
import pandas as pd

# Specifica il percorso del tuo file CSV
file_path = "etfs/mkt_values.csv"

# Carica i dati dal file CSV
df = pd.read_csv(file_path)

# Converti la colonna "Date" in un oggetto datetime
df['Date'] = pd.to_datetime(df['Date'])

# Estrai l'anno dalla colonna "Date" e crea una nuova colonna chiamata "Year"
df['Year'] = df['Date'].dt.year

# Raggruppa per anno e calcola la media della colonna "daily_return"
media_annua = df.groupby('Year')['daily_return'].mean()

# Mostra il risultato
print("Media annua della colonna 'daily_return':")
print(media_annua)
```

## HO STAMPATO LE MEDIE MENSILI

```python
import os
import pandas as pd

# Definisci la cartella contenente i file CSV
cartella = 'Data/etfs'

# Loop attraverso tutti i file nella cartella
for file in os.listdir(cartella):
    # Verifica se il file attuale ha l'estensione ".csv"
    if file.endswith(".csv"):
        # Crea il percorso completo del file
        percorso_file = os.path.join(cartella, file)
        # Carica il file CSV in un DataFrame
        df = pd.read_csv(percorso_file)
        # Converti la colonna 'Date' in formato datetime
        df['Date'] = pd.to_datetime(df['Date'])
        # Calcola la media dei rendimenti giornalieri per ogni mese
        media_mensile = df.groupby(df['Date'].dt.to_period('M'))['daily_return'].mean()
        # Crea una nuova colonna 'monthly_mean' nel DataFrame originale e assegna i valori delle medie mensili
        df['monthly_mean'] = df['Date'].dt.to_period('M').map(media_mensile)
        # Salva il DataFrame modificato nel file CSV originale
        df.to_csv(percorso_file, index=False)
```

## STAMPO SCARTO QUADRATICO MEDIO 

```python
import os
import pandas as pd

# Percorso della cartella 'etfs' che contiene i file CSV
cartella = 'Data/etfs'

# Ottenere la lista dei file nella cartella
elenco_file = os.listdir(cartella)

# Ciclo attraverso ogni file nella cartella
for file in elenco_file:
    # Costruire il percorso completo del file
    percorso_file = os.path.join(cartella, file)

    # Assicurarsi che il file sia un file CSV
    if file.endswith('.csv'):
        # Caricare il file CSV in un DataFrame
        df = pd.read_csv(percorso_file)

        # Calcolare la differenza tra 'daily_return' e 'monthly_mean' per ogni riga
        df['diff_daily_monthly'] = df['daily_return'] - df['monthly_mean']

        # Aggiungere i risultati nella colonna 'diff_daily_monthly' al DataFrame
        # e sovrascrivere il file CSV originale
        df.to_csv(percorso_file, index=False)

        print(f"Risultati aggiunti al file {file}")
```

## STAMPO LE 10 AZIENDE PIU' GROSSE DI ETFS

```python
import os
import pandas as pd

# Definisci la cartella
cartella = 'Data/etfs'

# Lista per contenere i nomi dei file e le rispettive somme di Volume
sommario_volume = []

# Scansiona tutti i file nella cartella
for filename in os.listdir(cartella):
    if filename.endswith(".csv") and filename != "mkt_values.csv":
        file_path = os.path.join(cartella, filename)
        
        # Carica il file CSV con pandas
        df = pd.read_csv(file_path)
        
        # Calcola la somma del Volume
        somma_volume = df['Volume'].sum()
        
        # Aggiungi il nome del file e la somma del Volume alla lista
        sommario_volume.append({'file': filename, 'somma_volume': somma_volume})

# Ordina la lista in base alla somma del Volume in ordine decrescente
sommario_volume = sorted(sommario_volume, key=lambda x: x['somma_volume'], reverse=True)

# Stampa i primi 10 file con la somma più alta
for entry in sommario_volume[:10]:
    print(f"File: {entry['file']}, Somma di Volume: {entry['somma_volume']}")
```

## CALCOLO SQM DELLE AZIENDE IN ETFS

```python
import os
import pandas as pd
import numpy as np

# Cartella contenente i file CSV
cartella = 'Data/etfs'

# Lista dei file da elaborare (prendi tutti i file nella cartella)
elenco_file = [file for file in os.listdir(cartella) if file.endswith('.csv')]

# Lista per memorizzare gli scarti quadratici medi mensili
scarti_quadratici_mensili = []

# Loop attraverso i file specificati
for file in elenco_file:
    # Crea il percorso completo al file
    percorso_file = os.path.join(cartella, file)

    # Carica il DataFrame da ciascun file CSV
    df = pd.read_csv(percorso_file)

    # Converti la colonna 'Date' in formato datetime
    df['Date'] = pd.to_datetime(df['Date'])

    # Raggruppa i dati per mese
    gruppo_mensile = df.groupby(df['Date'].dt.to_period("M"))

    # Calcola lo scarto quadratico medio per ogni mese
    for nome_mese, dati_mese in gruppo_mensile:
        scarto_quadratico_medio = np.sqrt(np.mean((dati_mese['daily_return'] - dati_mese['monthly_mean'])**2))
        scarti_quadratici_mensili.append({
            'ETF': file[:-4],  # Nome ETF senza l'estensione .csv
            'Mese': nome_mese.month,
            'Anno': nome_mese.year,
            'ScartoQuadraticoMedio': scarto_quadratico_medio
        })

# Visualizza i risultati
for risultato in scarti_quadratici_mensili:
    print(f"ETF: {risultato['ETF']}, Mese: {risultato['Mese']}, Anno: {risultato['Anno']}, Scarto Quadratico Medio: {risultato['ScartoQuadraticoMedio']}")
```
