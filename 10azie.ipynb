```python
import os
import pandas as pd
import numpy as np

# Cartella contenente i file CSV
cartella = 'Data/etfs'

# Lista dei file da elaborare
elenco_file = ['BKLN.csv', 'ARKK.csv', 'AGG.csv', 'DIA.csv', 'BND.csv', 'ASHR.csv', 'ACWI.csv', 'DXJ.csv', 'AMLP.csv', 'DBC.csv']

# Lista per memorizzare gli scarti quadratici medi mensili
scarti_quadratici_mensili = []

# Loop attraverso i file specificati
for file in elenco_file:
    # Crea il percorso completo al file
    percorso_file = os.path.join(cartella, file)

    # Carica il DataFrame da ciascun file CSV
    df = pd.read_csv(percorso_file)

    # Converti la colonna 'Date' in formato datetime
    df['Date'] = pd.to_datetime(df['Date'])

    # Raggruppa i dati per mese
    gruppo_mensile = df.groupby(df['Date'].dt.to_period("M"))

    # Calcola lo scarto quadratico medio per ogni mese
    for nome_mese, dati_mese in gruppo_mensile:
        scarto_quadratico_medio = np.sqrt(np.mean((dati_mese['daily_return'] - dati_mese['monthly_mean'])**2))
        scarti_quadratici_mensili.append({
            'ETF': file[:-4],  # Nome ETF senza l'estensione .csv
            'Mese': nome_mese.month,
            'Anno': nome_mese.year,
            'ScartoQuadraticoMedio': scarto_quadratico_medio
        })

# Visualizza i risultati
for risultato in scarti_quadratici_mensili:
    print(f"ETF: {risultato['ETF']}, Mese: {risultato['Mese']}, Anno: {risultato['Anno']}, Scarto Quadratico Medio: {risultato['ScartoQuadraticoMedio']}")
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Cartella contenente i file CSV
cartella_etfs = 'Data/etfs'

# Elenco delle aziende
aziende = ['BKLN', 'ARKK', 'AGG', 'DIA', 'BND', 'ASHR', 'ACWI', 'DXJ', 'AMLP', 'DBC']

# Creazione del grafico
plt.figure(figsize=(12, 8))

# Itera attraverso ogni azienda
for azienda in aziende:
    # Costruisci il percorso completo del file CSV
    file_path = os.path.join(cartella_etfs, f'{azienda}.csv')
    
    # Carica il file CSV e calcola la media dei rendimenti giornalieri per ogni anno
    df = pd.read_csv(file_path, parse_dates=['Date'])
    rendimenti_annuali_media = df.groupby(df['Date'].dt.year)['daily_return'].mean()
    
    # Disegna il grafico per ogni azienda
    plt.plot(rendimenti_annuali_media.index, rendimenti_annuali_media, label=azienda)

# Personalizzazione del grafico
plt.title('Rendimenti Annuali Medi delle Aziende')
plt.xlabel('Anno')
plt.ylabel('Rendimento Annuo Medio')
plt.legend()
plt.grid(True)

# Visualizza il grafico
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Cartella contenente i file CSV
cartella_etfs = 'Data/stocks'

# Elenco delle nuove aziende
aziende_nuove = ['AAPL', 'AMZN', 'BAC', 'AMD', 'CSCO', 'CMCSA', 'AAL', 'ABEV', 'BBD', 'C']

# Creazione del grafico
plt.figure(figsize=(12, 8))

# Itera attraverso ogni azienda
for azienda in aziende_nuove:
    # Costruisci il percorso completo del file CSV
    file_path = os.path.join(cartella_etfs, f'{azienda}.csv')
    
    # Carica il file CSV e calcola la media dei rendimenti giornalieri per ogni mese
    df = pd.read_csv(file_path, parse_dates=['Date'])
    rendimenti_mensili_media = df.groupby([df['Date'].dt.year, df['Date'].dt.month])['daily_return'].mean()
    
    # Calcola gli anni e i mesi come nuovi indici
    rendimenti_mensili_media.index = [f'{anno}-{mese}' for anno, mese in rendimenti_mensili_media.index]
    
    # Disegna il grafico per ogni azienda
    plt.plot(rendimenti_mensili_media.index, rendimenti_mensili_media, label=azienda)

# Personalizzazione del grafico
plt.title('Rendimenti Mensili Medi delle Nuove Aziende')
plt.xlabel('Anno-Mese')
plt.ylabel('Rendimento Mensile Medio')
plt.xticks(rotation=45)  # Ruota le etichette sull'asse delle ascisse per una migliore leggibilità
plt.legend()
plt.grid(True)

# Visualizza il grafico
plt.show()
```
